## 3. Realiza un cálculo de al menos dos medidas estadísticas descriptivas clave (por ejemplo, media, mediana, desviación estándar) utilizando NumPy. ¿Qué información te proporcionan estas medidas sobre los datos?
Las medidas estadísticas descriptivas nos permiten resumir la información de las variables numéricas del dataset. En este caso, aplicaremos las medidas de:

- Media (promedio): valor promedio de los puntajes, muestra la tendencia central.  
- Mediana: el valor central cuando los datos se ordenan, útil cuando hay valores atípicos.  
- Desviación estándar: mide cuánto se dispersan los datos respecto a la media.  

Aplicaremos estos cálculos en las tres columnas de puntajes:
- `math score` (Matemáticas)  
- `reading score` (Lectura)  
- `writing score` (Escritura)  

```python
# Importar librerías necesarias
import pandas as pd
import numpy as np

# Cargar el dataset 
df = pd.read_csv("StudentsPerformance.csv")

# Seleccionar las columnas numéricas
scores = df[["math score", "reading score", "writing score"]]

# Calcular medidas estadísticas
media = np.mean(scores, axis=0)
mediana = np.median(scores, axis=0)
desviacion = np.std(scores, axis=0)

print("Medidas estadísticas descriptivas")
print("Media:\n", media)
print("\nMediana:\n", mediana)
print("\nDesviación estándar:\n", desviacion)

```

## Interpretacion de los numeros
## Matemáticas: 
Media: 66.09
Mediana: 66
Desviación estándar: 15.16
Los estudiantes de matematicas se ubican, en promedio, en un puntaje medio (66/100), con bastante dispersión de más o menos 15 puntos alrededor del promedio.

## Lectura: 
Media: 66.17
Mediana: 70
Desviación estándar: 14.60
El rendimiento promedio es un poco mejor que en matemáticas. Los puntajes se concentran más alrededor del 70, con una variabilidad moderada.

## Escritura:
Media: 68.05
Mediana: 69
Desviación estándar: 15.20
El promedio de escritura es similar al de lectura, pero con una dispersión un poco mayor, indicando diferencias notables entre estudiantes que escriben muy bien y otros con dificultades.

En conclusión, los datos muestran que los estudiantes tienden a obtener mejores puntajes en lectura y escritura que en matemáticas. Además, la desviación estándar nos dice que no todos los estudiantes rinden igual, pues hay diferencias considerables entre los más y menos hábiles en cada materia.

## 6. Seleccione dos variables que podrían estar relacionadas dentro del contexto del problema. Convierta las series correspondientes a estas variables en vectores de NumPy e implemente una regresión lineal utilizando numpy.polyfit(). 

## Selección de variables
Se seleccionaron las variables reading score (lectura) y writing score (escritura) porque representan dos habilidades estrechamente relacionadas.  
La hipótesis es que los estudiantes con mejor puntaje en lectura también tienden a tener mejor puntaje en escritura.  
## Conversión a vectores NumPy
A continuación, se convierten las columnas del DataFrame en arreglos NumPy para trabajar con ellos en la regresión:

```python
import numpy as np

# Extraer las variables en vectores NumPy
x = df["reading score"].to_numpy()
y = df["writing score"].to_numpy()

print(x[:5])  # Primeros 5 valores de lectura
print(y[:5])  # Primeros 5 valores de escritura

# Implementación de la regresión lineal con numpy.polyfit()
# Ajustar la regresión lineal (grado 1)
coeficientes = np.polyfit(x, y, 1)

pendiente, intercepto = coeficientes
print("Pendiente:", pendiente)
print("Intercepto:", intercepto)

# Crear la recta de regresión
modelo = np.poly1d(coeficientes)

# Generar valores predichos
y_pred = modelo(x)

```

## Interpretación:
El significado de los resultados obtenidos, indican que, en promedio, por cada punto adicional en lectura, la calificación en escritura aumenta casi en 1 punto. La relación es muy cercana a una proporción 1 a 1, lo que refleja una dependencia lineal fuerte entre ambas competencias.

```python
#Visualización de la regresión
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(x, y, alpha=0.5, label="Datos reales")
plt.plot(x, y_pred, color="red", label="Recta de regresión")
plt.xlabel("Reading Score")
plt.ylabel("Writing Score")
plt.title("Relación entre Lectura y Escritura")
plt.legend()
plt.show()
```

El análisis de regresión lineal realizado entre las variables Lectura y Escritura evidencia una relación positiva y prácticamente proporcional, en la que los incrementos en la primera se reflejan casi de manera directa en la segunda. Esta fuerte correlación sugiere que ambas competencias se encuentran estrechamente ligadas y que las intervenciones educativas orientadas a fortalecer la comprensión lectora podrían impactar de forma significativa en la habilidad de escritura. En consecuencia, el modelo confirma la pertinencia de considerar estrategias integradas de enseñanza, donde el desarrollo de una competencia refuerce simultáneamente la otra.

## 7. Grafica las variables que has seleccionado usando un gráfico de dispersión y añade un gráfico lineal determinado por los coeficientes de la relación de regresión. 

```python
# Gráfico de dispersión con línea de regresión

import matplotlib.pyplot as plt
import numpy as np

# Seleccionamos las variables
x = df["reading score"].values
y = df["writing score"].values

# Calculamos la regresión lineal con polyfit
m, b = np.polyfit(x, y, 1)

# Creamos la figura
plt.figure(figsize=(8, 6))

# Gráfico de dispersión
plt.scatter(x, y, alpha=0.5, label="Datos reales", color="blue")

# Línea de regresión
plt.plot(x, m*x + b, color="red", linewidth=2, label=f"y = {m:.2f}x + {b:.2f}")

# Etiquetas y título
plt.xlabel("Puntaje en Lectura")
plt.ylabel("Puntaje en Escritura")
plt.title("Relación entre Puntajes de Lectura y Escritura")
plt.legend()
plt.grid(True)
plt.show()
```

Al analizar el gráfico de dispersión junto con la línea de regresión lineal, se observa una tendencia clara y positiva entre las variables puntaje en Lectura y puntaje en Escritura. La recta ajustada muestra una pendiente cercana a 1 (𝑚≈0.99), lo que indica que, en promedio, por cada punto adicional obtenido en Lectura, el estudiante también incrementa casi en un punto su resultado en Escritura.

Esta relación tan estrecha sugiere que ambas competencias están fuertemente asociadas. Tiene sentido, dado que la comprensión lectora facilita la construcción de ideas, la argumentación y la redacción de textos, lo cual se refleja directamente en un mejor desempeño en Escritura.

La nube de puntos se ajusta de manera consistente alrededor de la línea de regresión, lo que significa que la línea describe adecuadamente la tendencia general de los datos. Aunque siempre existen casos puntuales donde un estudiante puede rendir mejor en una habilidad que en otra, la correlación positiva indica que la mayoría de los estudiantes siguen este patrón.

En conclusión, se puede afirmar que el análisis respalda la hipótesis de que mejores habilidades en Lectura están asociadas a mejores resultados en Escritura, lo cual es relevante porque orienta a los docentes a fortalecer programas de comprensión lectora como estrategia indirecta para mejorar también la producción escrita.
